```python
# DROPING some columns -

covid_df.drop(["Sno", "Time", "ConfirmedIndianNational", "ConfirmedForeignNational"], inplace= True, axis= 1)
```

```python
covid_df.info()
```

```python
# Converting INTO DATETIME


covid_df["Date"] = pd.to_datetime(covid_df["Date"], format= "%d-%m-%Y")
```

```python
covid_df.info()
```

```python
# ACTIVE CASES -

covid_df["Active Cases"] = (covid_df["Confirmed"]  - (covid_df["Cured"] + covid_df["Deaths"]))
covid_df.tail(3)
```

```python
# PIVOT TABLE - (next 4-5 lines for pivot table)


statewise = pd.pivot_table(covid_df, values= ["Confirmed","Deaths","Cured"], index= "State/UnionTerritory", aggfunc= max)
```

```python
statewise["Recovery Rate"] = statewise["Cured"]*100/statewise["Confirmed"]
```

```python
statewise["Mortality Rate"] = statewise["Deaths"]*100/statewise["Confirmed"]
```

```python
statewise = statewise.sort_values(by= "Confirmed", ascending= False)
```

```python
statewise.style.background_gradient(cmap= "inferno")
```

```python
# TOP 10 ACTIVE CASES STATES -
# using GROUP BY - 
# The labels of x axis are over lapping so do this  -  

top_10_active = covid_df.groupby(by= "State/UnionTerritory").max()[["Active Cases","Date"]].sort_values(by= ["Active Cases"], ascending= False).reset_index()
fig = plt.figure(figsize=(16,9))
plt.title("Top 10 states with active cases in India", size= 25)

ax= sns.barplot(data= top_10_active.iloc[:10], y= "Active Cases" ,x= "State/UnionTerritory" , linewidth= 2)
plt.xlabel("Sates")
plt.ylabel("Total Active Cases")
plt.show()
```

```python
top_10_deaths_states = covid_df.groupby(by= "State/UnionTerritory").max()[["Deaths", "Date"]].sort_values(by= ["Deaths"], ascending= False).reset_index()

fig= plt.figure(figsize= (18,5))

plt.title("Top 10 states with most deaths", size= 25)

ax= sns.barplot(data= top_10_deaths_states.iloc[:12], y= "Deaths",x= "State/UnionTerritory" , linewidth= 2 , edgecolor= "black")

plt.xlabel("Sates")
plt.ylabel("Total Death Cases")
plt.show()
```

```python
# GROWTH TREND  - 

fig= plt.figure(figsize=(12,6))

ax= sns.lineplot(data= covid_df[covid_df["State/UnionTerritory"].
isin(["Maharashtra", "Karnataka", "Kerala", "Tamil Nadu", "Uttar Pradesh"])]
                 ,x= "Date", y="Active Cases", hue= "State/UnionTerritory")


ax.set_title("Top 5affected states in India", size= 16)
```

```python
# remove ROWS WHERE STATE = INDIA -- 

vaccine = vaccine_df[vaccine_df["State"] != "India"]
vaccine
```

```python
vaccine.rename(columns= {"Total Individuals Vaccinated" : "Total"}, inplace= True)
```

```python
# State with most vaccinations done -- 


max_vac = vaccine.groupby(by= "State")["Total"].sum().to_frame("Total")
max_vac = max_vac.sort_values("Total", ascending= False)[:6]
max_vac
```

```python
fig = plt.figure(figsize=(18,9))
plt.title("Top 5 Vaccinated States in India", size= 20)
x= sns.barplot(data= max_vac.iloc[:8], y= max_vac.Total, x= max_vac.index)
plt.xlabel("States")
plt.ylabel("Vaccination")
plt.show()
```
